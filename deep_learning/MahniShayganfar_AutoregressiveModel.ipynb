```python
import warnings
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
%matplotlib inline
from matplotlib.pylab import rcParams
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
import pmdarima as pm
from sklearn import metrics
from pmdarima.arima import auto_arima
warnings.filterwarnings('ignore')
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
flight_data = sns.load_dataset("flights")
flight_data.head()
flight_data.tail()
flight_data.shape
flight_data.describe()
flight_data.info()
```

```python
flight_data_pivot = flight_data.pivot('month', 'year', 'passengers')
flight_data_pivot
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(flight_data_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.show();
```

```python
month = pd.date_range('19490131', periods=144, freq='M')
month
```

```python
flight_data['datestamp'] = month
flight_data.head()
```

```python
flight_data.rename(columns = {'year':'Year', 'month':'Month', 'passengers':'Passengers', 'datestamp':'Datestamp'}, inplace = True)
flight_data
```

```python
data = flight_data.loc[:, ('Datestamp', 'Passengers')]
data.head()
data.describe()
```

```python
data.set_index('Datestamp', inplace=True)
data.head()
```

```python
plt.figure(figsize=(15,10))
plt.plot(data)
plt.xlabel('Time')
plt.ylabel('Passengers in 1000')
plt.title('# of Passengers over Time')
plt.show();
```

```python
decomposition = seasonal_decompose(data, model='multiplicative')
```

```python
plt.figure(figsize=(20,10))

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(221)
plt.plot(data, color='#00b8ff', label='Original')
plt.legend(loc='best')

plt.subplot(222)
plt.plot(trend, color='b', label='Trend')
plt.legend(loc='best')

plt.subplot(223)
plt.plot(seasonal, color='#ff00ff', label='Seasonality')
plt.legend(loc='best')

plt.subplot(224)
plt.plot(residual, color='y', label='Residuals')
plt.legend(loc='best')

plt.show();
```

```python
decomposition = seasonal_decompose(data, model='additive')
```

```python
plt.figure(figsize=(20,10))

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(221)
plt.plot(data, color='#00b8ff', label='Original')
plt.legend(loc='best')

plt.subplot(222)
plt.plot(trend, color='b', label='Trend')
plt.legend(loc='best')

plt.subplot(223)
plt.plot(seasonal, color='#ff00ff', label='Seasonality')
plt.legend(loc='best')

plt.subplot(224)
plt.plot(residual, color='y', label='Residuals')
plt.legend(loc='best')

plt.show();
```

Test for the stationarity of the time series using **Augmented Dickey Fuller Test**. Since p value is higher than alpha, we cant reject the null hypothesis (Series is non stationary). Therefore, we would need to use the Integrated (I) term of the time series in a good model.

```python
from pmdarima.arima.stationarity import ADFTest
adf_test = ADFTest(alpha=0.05)
adf_test.should_diff(data)
```

```python
train, test = data[:100], data[100:]
train.shape
test.shape
```

```python
plt.plot(train)
plt.plot(test)
plt.show();
```

```python
Arima_model = auto_arima(train, start_p=1, start_q=1, max_p=8, max_q=8, 
                         start_P=1, start_Q=1, max_P=8, max_Q=8,
                         m=12, seasonal=True, trace=True, d=1, D=1,
                         error_action='warn', suppress_warnings=True,
                         stepwise=True, random_state=20, n_fits=30)
Arima_model.summary()
```

#### Use the best model to make predictions about the Test data.

```python
prediction = pd.DataFrame(Arima_model.predict(n_periods=44), index=test.index)

prediction.columns = ['predicted_passengers']
```

```python
prediction
```

```python
plt.figure(figsize=(15,10))
plt.plot(train, label='Training')
plt.plot(test, label='Test')
plt.plot(prediction, label='Predicted')
plt.legend(loc='upper center')
plt.show();
```

```python
test['Predicted_Passengers'] = prediction
test['Error'] = test['Passengers'] - test['Predicted_Passengers']
test
```

```python
metrics.mean_absolute_error(test.Passengers, test.Predicted_Passengers)
```

```python
metrics.mean_squared_error(test.Passengers, test.Predicted_Passengers)
```

```python
metrics.median_absolute_error(test.Passengers, test.Predicted_Passengers)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)
plt.plot(test.Error, color='#ff33cc')
plt.title('Error Distribution Over Time')

plt.subplot(122)
scipy.stats.probplot(test.Error, plot=plt)
plt.show();

plt.figure(figsize=(20,10))
pm.autocorr_plot(test.Error)
plt.show();
```
